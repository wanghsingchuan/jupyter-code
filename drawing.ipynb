```python
#產生 sin wave 圖形
import numpy as np
import matplotlib.pyplot as py
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
#對齊---------------
py.xlim(0,360)
py.ylim(-1.2,1.2)
py.title("sin wave")
#對齊---------------
py.scatter(x,y)
py.show()
```

```python
#產生 sin cos wave圖形
import numpy as np
import matplotlib.pyplot as py
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
#對齊---------------
py.xlim(0,360)
py.ylim(-1.2,1.2)
py.title("sin wave")
#對齊---------------
py.scatter(x,y, color="blue", label="sin wave")
py.scatter(x,z, color="red", label="cos wave")
py.legend()
py.show()
```

```python
#產生圖表
import numpy as np
import matplotlib.pyplot as py
x=[1,2,3,4]
y=[3,5,7,2]
py.bar(x,y, label="bar wave")
py.legend()
py.show()
```

```python
from sklearn import datasets, cluster, metrics
import matplotlib.pyplot as pt 
iris=datasets.load_iris()
iris_Data=iris.data

ks=range(2,11)
silhouette_avgs=[]

for n in ks :
    
    result=cluster.KMeans(n_clusters=n).fit(iris_Data)
    silhouette_avg=metrics.silhouette_score(iris_Data, result.labels_)
    silhouette_avgs.append(silhouette_avg)
pt.bar(ks, silhouette_avgs)
pt.show()
```

```python
from sklearn import datasets, cluster, metrics
from sklearn import svm
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt

digits=datasets.load_digits()

X_train, X_test, y_train, y_test, images_train, images_test = train_test_split(digits.data, digits.target, digits.images, test_size=0.25, random_state=42)

svc_model = svm.SVC(gamma=0.001, C=100., kernel='linear')
svc_model.fit(X_train, y_train)
predicted = svc_model.predict(X_test)

images_and_predictions = list(zip(images_test, predicted))
for index, (image, prediction) in enumerate(images_and_predictions[:4]):
    plt.subplot(1, 4, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.binary)
    plt.title('Predicted: ' +str(prediction))
plt.show
```
